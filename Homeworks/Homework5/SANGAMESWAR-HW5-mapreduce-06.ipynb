```python
#Reference
#https://www.lynda.com/Business-Intelligence-tutorials/Text-mining-Python/475936/529759-4.html


% matplotlib inline

import codecs
import re
import copy
import collections

import numpy as np
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import WordPunctTokenizer
from __future__ import division
from nltk.corpus import stopwords

with codecs.open("SherlockHolmesBooks/Doc1.txt", "r", encoding="latin-1") as f:
    Doc_1 = f.read()
with codecs.open("SherlockHolmesBooks/Doc2.txt", "r", encoding="latin-1") as f:
    Doc_2 = f.read()
with codecs.open("SherlockHolmesBooks/Doc3.txt", "r", encoding="latin-1") as f:
    Doc_3 = f.read()
with codecs.open("SherlockHolmesBooks/Doc4.txt", "r", encoding="latin-1") as f:
    Doc_4 = f.read()
with codecs.open("SherlockHolmesBooks/Doc5.txt", "r", encoding="latin-1") as f:
    Doc_5 = f.read()
    
esw = stopwords.words('english')
esw.append("would")

word_pattern = re.compile("^\w+$")

def get_text_counter(text):
    tokens = WordPunctTokenizer().tokenize(PorterStemmer().stem(text))
    tokens = list(map(lambda x: x.lower(), tokens))
    tokens = [token for token in tokens if re.match(word_pattern, token) and token not in esw]
    return collections.Counter(tokens), len(tokens)

def make_df(counter, size):
    abs_freq = np.array([el[1] for el in counter])
    rel_freq = abs_freq / size
    index = [el[0] for el in counter]
    df = pd.DataFrame(data=np.array([rel_freq]).T, index=index, columns=["Frequency"])
    df.index.name = "Most common words"
    return df
```

```python
d1_counter, d1_size = get_text_counter(Doc_1)
make_df(d1_counter.most_common(20), d1_size)
```

```python
d2_counter, d2_size = get_text_counter(Doc_2)
make_df(d2_counter.most_common(20), d2_size)
```

```python
d3_counter, d3_size = get_text_counter(Doc_3)
make_df(d3_counter.most_common(20), d3_size)
```
